## **Matplotlib Basics**

Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations.


#### Today, we wil try out a common module `matplotlib.pyplot`

```python
import matplotlib.pyplot as plt

# Create the data for plotting
labels = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
```

##### Overall customization

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 10
plt.rcParams['axes.titlesize'] = 12
```

#### 1. Bar Chart

```python
plt.figure(figsize=(6,4))
plt.bar(labels, values, color='skyblue')
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

#### 2. Line Chart

```python
plt.figure(figsize=(6,4))
plt.plot(labels, values, marker='s', linestyle='-', linewidth=2, markersize=7, color='green')
# Linestyle: '-' (solid), '--' (dashed), '-.' (dash-dot), ':' (dotted)
# Markers: 'o' (circle), 's' (square), 'D' (diamond), '^' (triangle)
plt.title("Line Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

#### 3. Pie Chart

```python
plt.figure(figsize=(4,4))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
plt.title("Pie Chart")
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular.
plt.show()
```

#### 4. Scatter Plot

```python
x = [5, 7, 8, 5, 6, 7, 8, 9]
y = [10, 12, 15, 10, 13, 15, 14, 16]

plt.figure(figsize=(6,4))
plt.scatter(x, y, color='purple')
plt.title("Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")

# Save the figure
plt.savefig("scatter_plot.png")
plt.show()
```
