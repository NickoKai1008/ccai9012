### **Class & Methods for Object-Oriented Programming (OOP)**

# Let's try to define a *human* in Python

<img src="../../docs/figs/w3/human.jpg" width="50%">

`__init__` is a constructor. Its main purpose is to set up(initialize) the object’s attributes. 

`self` refers to the specific object instance.

```python
 # --- Class Attributes (Shared by the species) ---
class Human:
    species = "Homo Sapien"
    legs = 2
    eyes = 2
    tail = 0
   
    # --- Instance Attributes (Unique to the person) ---   
    def __init__(self, name, haircolor, age, height, faculty):
        self.name = name
        self.haircolor = haircolor
        self.age = age
        self.height = height
        self.faculty = faculty
```

## Congratulations, you have created the class - *human*
## Now, who is a *human*?

```python
# Creating the Instances/Objects
uid_1 = Human("Joseph", "black", 25, 180, "Architecture")
uid_2 = Human("Joshua", "brown", 22, 175, "Medicine")

print(f"Hi I am {uid_1.name}, and I am a {uid_1.species}")
```

```python
# Methods:
# Method 1
def selfintro(self):
    return f"Hi, my name is {self.name} from the Faculty of {self.faculty}. I am {self.age} years old and my hair is {self.haircolor}."
# Method 2 
def action(self, action):
    return f"{self.name} is {action} now."
```

## Introduce yourself!

```python
# Calling Method 1
print(selfintro(uid_1))
print(selfintro(uid_2))

# Calling Method 2
print(action(uid_1, "jumping"))
```

## Dyeing my hair ^^

```python
# Update instance attributes
uid_1.haircolor = "red"
print(selfintro(uid_1))
```

## More info about me...

```python
# Add instance attributes
uid_1.weight = "200kg"
# Adding new info to Method 1
print(selfintro(uid_1) + f" I weigh {uid_1.weight}.")
```

```python
# this will run into an error because uid_2 has no weight
print(selfintro(uid_2) + f" I weigh {uid_2.weight}.")
```

## Better delete it ;)

```python
# this will run into an error because uid_1's weight is deleted
del uid_1.weight
print(uid_1.weight)
```

## Genetic evolution?!

```python
# Update class attributes
Human.tail = 1 # changes apply to all instances
print(f"{uid_1.name} has {uid_1.tail} tail!")
```
