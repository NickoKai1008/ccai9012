# Python 102: Classes and Methods

In this tutorial, we will explore **methods** and **classes**, which are the foundation of object-oriented programming (OOP) in Python.

We'll cover:
1. How to define a class
2. How to create objects (instances)
3. How to define and use methods
4. How to work with class and instance attributes

## 1. Defining a Simple Class
A **class** is a blueprint for creating objects. Each object can have its own data (**attributes**) and can perform actions through **methods**.

```python
# Define a class called Dog
class Dog:
    # Class attribute: shared by all instances
    species = "Canis familiaris"

    # Constructor method: called when creating a new object
    def __init__(self, name, age):
        self.name = name  # instance attribute
        self.age = age    # instance attribute

    # Method to describe the dog
    def description(self):
        return f"{self.name} is {self.age} years old"

    # Method to make the dog "speak"
    def speak(self, sound):
        return f"{self.name} says {sound}"
```

## 2. Creating Objects (Instances)

We can create multiple objects (instances) from the same class. Each instance can have its own attributes.

```python
# Create two Dog objects
dog1 = Dog("Buddy", 3)
dog2 = Dog("Lucy", 5)

# Access instance attributes
print(dog1.name)  # Buddy
print(dog2.age)   # 5

# Call methods
print(dog1.description())  # Buddy is 3 years old
print(dog2.speak("Woof!")) # Lucy says Woof!
```

## 3. Modifying Object Attributes

Object attributes can be changed after creation:

```python
# Update dog1's age
dog1.age = 4
print(dog1.description())  # Buddy is 4 years old

# Add a new attribute dynamically
dog1.color = "brown"
print(f"{dog1.name} is {dog1.color}")  # Buddy is brown
```

## 4. Class Attributes

Class attributes are shared by all instances:

```python
# Access class attribute
print(dog1.species)  # Canis familiaris
print(dog2.species)  # Canis familiaris

# Change class attribute for all instances
Dog.species = "Dog"
print(dog1.species)  # Dog
print(dog2.species)  # Dog
```

## 5. Summary

- A **class** is a blueprint for objects.
- An **object** (or instance) is a specific instance of a class.
- **Attributes** store data about the object.
- **Methods** define actions the object can perform.
- `self` refers to the specific object instance.
- Class attributes are shared by all objects, while instance attributes are unique to each object.
