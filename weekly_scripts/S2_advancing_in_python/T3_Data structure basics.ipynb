### **Data Structure Basics**

# *[List]*
### ordered, changeable, allows duplicates
`list = [item, item, item]`

<img src="../../docs/figs/w3/list.png" width="30%">

- Access elements

```python
fruits = ["apple", "banana", "cherry"]

# Access item(s) via index
print(fruits[0])
print(fruits[0:2])
# Access index via item
print(fruits.index("apple"))
```

- Change item

```python
fruits[1] = "blueberry"
print("Changed list:", fruits)
```

- Insert item

```python
fruits.insert(2, "watermelon")
print("After add:", fruits)
```

- Add an item to the end

```python
fruits.append("pineapple")
print("After append:", fruits)
```

- Remove items

```python
# use reomove
fruits.remove("apple")
print("After remove:", fruits)

# use del
del fruits[2]
print("After del:", fruits)
```

- Extract items

```python
extract_item = fruits.pop(1)
print("Extracted item:", extract_item)
print("After pop:", fruits)
```

- Get list length

```python
print("Number of fruits left:", len(fruits))
```

# *(Tuple)*
### ordered and unchangeable/ immutable
`tuple = (item, item, item)`

```python
fruits = ("apple", "banana", "cherry")
print(fruits[0])
```

# *{Dictionary}*
### A dictionary stores values with keys
`dictionary = {key:value, key:value, key:value}`

```python
# To get a value from a dictionary, use the key inside square brackets `[]` or with `.get()`.
person = {"name": "Joseph", "age": 22}
print("Name:", person["name"])   # Access value by key
print("Age:", person.get("age")) # Another way to get value
```

- Add a new key-value pair

```python
person["city"] = "New York"
print("City:", person["city"])
```

- Change an existing value

```python
person["age"] = 23
print("Age:", person["age"])
```
