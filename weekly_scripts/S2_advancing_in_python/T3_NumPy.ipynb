## **NumPy Basics**

NumPy (Numerical Python) is a Python library used for fast numerical operations.

- It allows you to work with large arrays and matrices.
- It's much faster and more efficient than Python lists for math operations.
- It is the foundation for many scientific libraries in Python (like Pandas, Scikit-learn, etc).

#### Why not just use lists?
- Lists are flexible, but slow for large-scale calculations.
- NumPy arrays use less memory and support powerful vectorized operations (doing math on many values at once).

## 1. Array (list)

#### Array via `array`

```python
# Import numpy
import numpy as np

# Create a NumPy array from a list
a = np.array([1, 2, 3, 4])
print("Array:", a)
```

#### Array via `arange`

```python
# range(start,stop,step)
for range_a in range(1,10,2):
    print(range_a)
```

```python
# np.arange(start,stop,step)
range_a = np.arange(0,9,2)
print("Range array:", range_a)
```

#### Basic statistics

```python
print("Mean:", np.mean(range_a))
print("Median:", np.median(range_a))
print("Sum:", np.sum(range_a))
print("Max:", np.max(range_a))
print("Min:", np.min(range_a))
print("+10:", range_a + 10)
print("x10:", range_a * 10)
```

## 2. Matrix (table)

#### Matrix via `array`

```python
# 3x3 matrix
matrix = np.array([[1, 2, 3],[1, 2, 3],[1, 2, 3]])

print(matrix)
```

#### Matrix via `arange`

```python
# 3x3 matrix
matrix = np.arange(9).reshape(3, 3)

print(matrix)
```

```python
# 3x3 matrix
matrix = np.arange(0,17,2).reshape(3, 3)

print(matrix)
```

## Shape and dimension

```python
data = np.arange(120).reshape(2,3,4,5)

# Data type (dtype): tells what kind of data the array holds
print("Data type:", data.dtype)

# Number of dimensions (ndim): tells how many axes or directions the array has
print("Number of dimensions:", data.ndim)

# Shape: size of each dimension (length along each axis)
print("Shape:", data.shape)
```
