## **Pandas Basics**

Pandas is a powerful Python library for working with tabular data (data with labels/headers).

- It helps you read, organize, and analyze data easily.
- Main data structures are:
  - *Series*: a 1D labeled array
  - *DataFrame*: a 2D labeled table

## 1. Series

```python
import pandas as pd

# Create a Series (1D labeled array)
s = pd.Series([10, 20, 30, 40], index=["a", "b", "c", "d"])
s
```

```python
# Access value by position (index)
print("\nValue at index 0:", s.iloc[0])  # 10

# Access value by label (label)
print("Value with label 'c':", s.loc["c"])  # 30
```

## 2. DataFrame

```python
# Create a dictionary
# It looks like a rotated table
data = {
    "Name": ["Alice", "Bob", "Charlie"], 
    "Age": [25, 30, 35],
    "City": ["New York", "Paris", "London"]
}

# Turn it to a DataFrame
df = pd.DataFrame(data)
df # display in jupyter notebook
```

## 3. Accessing Data

#### Access column(s)

```python
# Access column
df["Name"]
```

```python
# Access columns
df[["Name", "Age"]]
```

#### Access row(s)

```python
# Access row
df.iloc[0]
```

```python
# Access rows
df.iloc[[0,2]]
```

```python
# Access rows with range
df.iloc[0:2]
```

#### Access a value (cell)

```python
data = {
    "Name": ["Alice", "Bob", "Charlie"], 
    "Age": [25, 30, 35],
    "City": ["New York", "Paris", "London"]
}

df = pd.DataFrame(data)
df
```

```python
# Using label-based .loc
print("Value at row 0, column 'Age':", df.loc[0, "Age"])  

# Using position-based .iloc
print("Value at row 1, column 'City':", df.iloc[1, 2])       
```

## 4. Operating Data

#### Basic statistics for a column

```python
print(df["Age"].mean())
print(df["Age"].median())
print(df["Age"].max())
print(df["Age"].min())
```

#### Filter rows

```python
# Filter where Age > 28
df[df["Age"] > 28]  
# inner df[] evaluates which rows meet the condition
# outer df[] extracts the successful rows from the main DataFrame
```

#### Add a new column

```python
df["Salary"] = [70000, 80000, 90000]
df
```

#### Add a new row

```python
df.loc[3] = ["David", 28, "Berlin", 60000]
df
```

## 5. Export & Import Data

#### Saving DataFrame to CSV

```python
df.to_csv("output.csv", index=False)  # index=False means row numbers are not saved
print("DataFrame saved to 'output.csv'")
```

#### Reading a CSV file into a DataFrame

```python
df = pd.read_csv("output.csv") 
df.head() # showing the fisrt few ros of the dataframe
```
