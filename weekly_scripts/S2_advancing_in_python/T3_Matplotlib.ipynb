## **Matplotlib Basics**

Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations.


#### Today, we wil try out a common module `matplotlib.pyplot`

#### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Overall customization

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 10
```

#### Load CSV

```python
df = pd.read_csv('data.csv')
```

## 1. Line Chart

```python
# Extract the first column
Lilo = df.iloc[:, 0] # [all rows, first column]

# Create the plot
plt.figure(figsize=(5, 2))
plt.plot(Lilo, color='blue', linewidth=1)

# Add labels and title
plt.title('Happiness level of Lilo', fontsize=12)
plt.xlabel('Days')
plt.ylabel('Happiness level')
plt.grid(linestyle='--', alpha=0.6) # alpha = opacity

plt.show()
```

## 5D!

```python
fig = plt.figure(figsize=(5,5)) # creates the figure (container for everything)
ax = fig.add_subplot(111, projection='3d') # ax = axes = plot/subplot inside figure

# Set projection and rotation
ax.set_proj_type('ortho') # orthographic projection
ax.view_init(elev=30, azim=-60) # Adjust these numbers to rotate

# Mapping the dimensions
# s = size, c = color
scatter = ax.scatter(
    df.iloc[:, 0],        # col1 as X
    df.iloc[:, 1],        # col2 as Y
    df.iloc[:, 2],        # col3 as Z
    s=df.iloc[:, 3] * 3,  # col4 as size, s = size (x3 to make it visible)
    c=df.iloc[:, 4],      # col5 as color, c = color
    cmap='viridis',       # cmap = color scheme
    alpha=0.6,            # alpha = opacity
    edgecolors='w'        # w = white
)

# Add a color bar
# scatter to call the plot, pad = margin, it knows we are refering to col5 as we associated color with it before
# ax=ax to associate the colorbar to the subplot
cbar = fig.colorbar(scatter, ax=ax, shrink=0.3, aspect=20, location='right', pad=0.2) 
cbar.set_label("Cobra")

ax.set_xlabel("Lilo", labelpad=5)   # labelpad = margin
ax.set_ylabel("Nani", labelpad=10)
ax.set_zlabel("Jumbo", labelpad=2)
plt.title("5D Data Visualization", fontsize=12)

plt.show()
```
