First, lets go through some Python concepts.
### 1. Variables and Data Types

**string**: `player_name` Stores text data. Using `.title()` is a string method that capitalizes the first letter of every word.

**integer**: `stamina` Stores whole numbers. It tracks the player's energy state.

**boolean**: True or False. These are "on/off" switches.

```python
#Print & Strings
print("You wake up in a locked room. It smells like old pizza.")
```

```python
#Variables & Inputs
player_name = input("What is your name, survivor of CCAI9012? ")
player_name = player_name.title() 
print("Good luck, " + player_name)
```

### 2. Import

Python doesn't include everything by default. To use specialized tools like randomization, you must import the `random` module. This gives you access to *randint(1,10)*, which generates a random integer between those two bounds.

### 3. String Formatting

The `f"..."` in your `input()` is a formatted string literal. It allows you to inject variables (like your random numbers/variables) directly into the text by wrapping them in curly braces {}.

```python
#Import, numbers & format
import random
num_1 = random.randint(1,10)
num_2 = random.randint(1,10)
num_3 = random.randint(1,10)
code_guess = input(f"A keypad flashes. What is {num_1} + {num_2} * {num_3}? ")
```

### 4. Control FLow(If-Else)

This is how your program makes decisions.

You have an `if` inside an `if`, it is called a *"nested-if"*. The first check `.isdigit()` ensures the program doesn't crash if the user enters a string. Once the input is validated, the inner `if` verifies the mathematical result. If either condition fails, the `else` block handles the fallout.

### 5. Comparison Operators

**`==` (Equal to)**: Checks if values match.

**`!=` (Not equal to)**: Checks if values are different.

**`<` `>` `>=` `<=`**: Checks numerical boundaries.

```python
#If-else & Comparison
if code_guess.isdigit(): # Is int?
    if int(code_guess) == num_1 + num_2 * num_3: # Is int correct?
        print("The door clicks open!")
    else:
        print("WRONG. A loud alarm starts blaring!")
else:
    print("WRONG. A loud alarm starts blaring!")
```

### 6. Loops

The `while` loop creates a repetition based on a condition.

**Condition**: -3 < stamina < 5 is a shorthand way of saying "as long as stamina is greater than -3 and less than 5."

**Termination**: The loop only stops when the condition becomes False (either the player wins or loses). You can also use `break` to stop the loop when a certain condition is fulfilled.

```python
#Loops
stamina = 0
while -3 < stamina < 5:
    action = input("Type 'run' to keep going: ")
    
    if action != "run": 
        print("You tripped!")
        stamina = stamina - 1
    else:
        print("You're getting closer...")
        stamina = stamina + 1
    
if stamina >= 5:
    print("🎉 You made it out alive!")
else:
    print("💀 You got caught...Game over!")
```

### 7. Functions

**definition**: `def play_keypad()` defines a function, tells Python to store these instructions for later use.

**return**: `return True` return sends an answer back to the part of the program. It's the "output" of your function.

**call**: `play_keypad()` executing the stored instructions. Allows you to run the same code multiple times without rewriting it.

## Now, lets put everything together! 

```python
#Print & Strings
print("You wake up in a locked room. It smells like old pizza.")

#Variables & Inputs
player_name = input("What is your name, survivor? ")
player_name = player_name.title() 
print("Good luck, " + player_name)

#Import, Functions & format
stamina = 0

import random
def play_keypad():
    num_1 = random.randint(1,10)
    num_2 = random.randint(1,10)
    num_3 = random.randint(1,10)
    code_guess = input(f"A keypad flashes. What is {num_1} + {num_2} * {num_3}? ")

    #If-else & Comparison
    if code_guess.isdigit(): # Is int?
        if int(code_guess) == num_1 + num_2 * num_3: # Is int correct?
            return True
        else:
            return False
    else:
        return False

if play_keypad() == True: #1st time
    print("The door clicks open! Another door appears...")
else:
    print("WRONG. A loud alarm starts blaring! Please try again.")
    stamina = stamina - 1

if play_keypad() == True: #2nd time
    print("The door clicks open!")
else:
    print("WRONG. In panic, you bump the door open.")
    stamina = stamina - 1

#Loops
while -3 < stamina < 5:
    action = input("Type 'run' to keep going: ")
    
    if action != "run": 
        print("You tripped!")
        stamina = stamina - 1
    else:
        print("You're getting closer...")
        stamina = stamina + 1
    
if stamina >= 5:
    print("🎉 You made it out alive!")
else:
    print("💀 You got caught...Game over!")
```
